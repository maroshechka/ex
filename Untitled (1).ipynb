# 4 Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? Не стоит забывать, что внутри месяца может быть не целое количество недель. Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть.

```python
orders
```

```python
# создаем столбец с годом и месяцем заказа
orders['order_approved_month'] = orders.order_approved_at.dt.strftime('%Y-%m')
```

```python
orders
```

```python
# определяем количество дней в месяце и считаем кол-во недель в месяце
orders['count_day_weeks'] = (orders.order_approved_at.dt.days_in_month / 7).
```

```python
orders
```

```python
orders_count = orders \
    .groupby(['customer_id', 'order_approved_month', 'count_day_weeks'], as_index=False) \
    .agg({'order_id': 'count'}) \
    .rename(columns={'order_id': 'orders_count'}) \
    .groupby('customer_id') \
    .agg({'orders_count': 'mean'}) \
    .rename(columns={'orders_count': 'orders_mean'})


```

```python
orders_count
```

```python
orders_count = orders_count.merge(orders[['customer_id', 'count_day_weeks']], on='customer_id', how='left')
orders_count['mean_orders'] = orders_count['orders_mean'] / orders_count['count_day_weeks']

```

```python
orders_count
```

# ВЫВОД: в дф orders_count представленно среднее кол-во покупок в месяц в столбце mean_orders

# 5.1. Выполните когортный анализ пользователей.

```python
df_cogort = pd.merge(orders, order_items, on='order_id', how='left')
```

```python
df_cogort
```

```python
first_order_month = df_cogort.groupby('customer_id')['order_approved_month'].min()
first_order_month.name = 'first_month' #Определение месяца первой покупки каждого пользователя
```

```python
df_cogort = df_cogort.join(first_order_month, on='customer_id')
```

```python
df_cogort
```

```python
df_cogort = df_cogort.groupby(['first_month', 'order_purchase_timestamp']) \
              .agg({'customer_id': 'nunique'}).rename(columns={'customer_id': 'num_of_users'})
```

```python
df_cogort
```

```python
df_cogort['cohort_period'] = df_cogort.groupby('first_month').cumcount()
df_cogort = df_cogort.reset_index()
```

```python
df_cogort
```

```python
cohort_table = df_cogort.pivot(columns = 'cohort_period', index = 'first_month', values= 'num_of_users')
cohort_table = cohort_table.divide(cohort_table.iloc[:,0],axis=0)
cohort_table
```
